```python
c=0
for product_link in products:
    c=c+1
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-notifications")
    options.add_argument("--disable-popup-blocking");
    options.add_argument("test-type")
    driver = webdriver.Chrome('/Users/akssingh/Downloads/PS_TWM_shadow-main/chromedriver', options = options)
    driver.get(product_link)
    time.sleep(5)#sleep_between_interactions
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)#sleep_between_interactions 

    try :
        #WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".agegate-closeX"))).click()
        close_btn = driver.find_element_by_id('agegate-closeX')
        if close_btn.is_displayed() :
            close_btn.click()
    except :
        print("No popup")
    
    
    w=driver.find_elements_by_xpath('//h2[@class="productTitle__3XDd9UVh"]')
    
    
    # all reviews
    button_class = '.bv-content-btn.bv-content-btn-pages.bv-content-btn-pages-load-more.bv-focusable'
    while True:
        try :
            driver.find_element_by_css_selector(button_class).click()
        except :
            break
    while True:
        try :
            driver.find_element_by_css_selector(button_class).click()
        except :
            break

    review_class = '.bv-content-list.bv-content-list-reviews'
    review_div = driver.find_element_by_css_selector(review_class)
    review_class = ".bv-content-item-author-profile-offset.bv-content-item-author-profile-offset-on"
    reviews = review_div.find_elements_by_css_selector(review_class)
    total_reviews = len(reviews)
    print("Total reviews: ", total_reviews)
    

    for i in range(0,total_reviews) :
        
        try :
            title = reviews[i].find_element_by_class_name('bv-content-title').text
        except :
            title = ""
            pass
        
        
        try :
            review_txt = reviews[i].find_element_by_class_name('bv-content-summary-body-text').text
            review_txt = review_txt.replace('\n' , ' ')
        except :
            review_txt = ""
        
        data['Product name'].append(w[0].text.replace('\n',' '))
        data['Reviews'].append(review_txt)
        data['Reviews_title'].append(title)

    driver.close()
    print("Product", str(c) , data['Product name'][-1] , "--> done")
    print('--------------------')
    time.sleep(50)
```
