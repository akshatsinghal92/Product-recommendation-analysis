```python
c=0
for product_link in products:
    c=c+1
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-notifications")
    options.add_argument("--disable-popup-blocking");
    options.add_argument("test-type")
    driver = webdriver.Chrome('/Users/akssingh/Downloads/chromedriver', options = options)
    driver.get(product_link)
    time.sleep(5)#sleep_between_interactions
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(8)#sleep_between_interactions 

    try :
        #WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".agegate-closeX"))).click()
        close_btn = driver.find_element_by_id('agegate-closeX')
        if close_btn.is_displayed() :
            close_btn.click()
    except :
        print("No popup")
    
    rat_cnt = int(driver.find_element_by_class_name('bv_numReviews_text').text[1:-1])
    if rat_cnt == 0:
        print("Product",str(c) , driver.find_elements_by_xpath('//h2[@class="productTitle__3XDd9UVh"]')[0].text.replace('\n',' ') , "--> SKIPPED")
        driver.close()    
        print('--------------------')
        time.sleep(45)
        continue
    
    data['Rating_count'].append(rat_cnt)
    
    w=driver.find_elements_by_xpath('//h2[@class="productTitle__3XDd9UVh"]')
    data['Product name'].append(w[0].text.replace('\n',' '))
    
    
    p=driver.find_elements_by_xpath('//div[@class="priceContainer__3IL5qTqJ"]')
    try:
        data['Price_per_unit'].append(float(p[0].text[1:6]))
    except:
        data['Price_per_unit'].append(float(p[0].text.replace(',','')[1:9]))
    
    
    s=driver.find_elements_by_xpath('//div[@class="productLabel__1GmVDiEr"]')
    for i in range(4,9):
        try:
            data['Similar_product_'+str(i-3)].append(s[i].text.replace('\n',' '))
        except:
            data['Similar_product_'+str(i-3)].append(np.nan)
        
    sp=driver.find_elements_by_xpath('//div[@class="productRatingAndPrice__1L5MGfhS"]')
    for i in range(4,9):
        try:
            data['Similar_product_price_'+str(i-3)].append(float(sp[i].text[-5:]))
        except:
            data['Similar_product_price_'+str(i-3)].append(np.nan)

        
    #rating=driver.find_elements_by_xpath('//div[@class="css-5aryqg"]')
    try:
        data['Rating'].append(float(driver.find_element_by_class_name('css-5aryqg').text[0:3]))
    except:
        data['Rating'].append(float(driver.find_element_by_css_selector('.bv_avgRating_component_container.notranslate').text[0:3]))
        
    
    

    # rating dist
    element_to_hover_over = driver.find_element_by_css_selector('.bv_avgRating_component_container.notranslate')
    hover = ActionChains(driver).move_to_element(element_to_hover_over)
    hover.perform()
    for i in range(0,5):
        key = driver.find_elements_by_class_name('bv_histogram_row_prefix')[i].text+' star count'
        val = driver.find_elements_by_class_name('bv_histogram_row_rating')[i].text
        data[key].append(int(val))

 
    #review - pros & cons
    jim_beam={}
    jim_beam['Pros']={}

    try:
        pros = driver.find_element_by_class_name('css-mg427s')
        pros = pros.find_elements_by_tag_name("li")
        for item in pros:
            txt = item.find_element_by_class_name('css-w99yow').text.split(' ')
            key = txt[0 : len(txt) - 2] # type of sentiment
            key = ' '.join(key)
            val = txt[-2:] # last two elems : num & review
            val = ' '.join(val)
            jim_beam['Pros'][key] = val   
        data['Pros'].append(jim_beam['Pros'])
    except:
        data['Pros'].append(np.nan)


    product={}
    product['Cons'] = {}

    try :
        cons = driver.find_elements_by_class_name('css-mg427s')[1]
        cons = cons.find_elements_by_tag_name("li")
        for item in cons:
            txt = item.find_element_by_class_name('css-w99yow').text.split(' ')
            key = txt[0 : len(txt) - 2] # type of sentiment
            key = ' '.join(key)
            val = txt[-2:] # last two elems : num & review
            val = ' '.join(val)
            product['Cons'][key] = val
        data['Cons'].append(product['Cons'])
    except:
        data['Cons'].append(np.nan)

    # users recommedning product
    try:
        data['product recommendation'].append(driver.find_element_by_class_name('css-1tpw5di').text)
        data['recommendation_count'].append(driver.find_element_by_class_name('css-66vicj').text.split(' ')[0]) #int
        data['recommendation_percent'].append(driver.find_element_by_class_name('css-66vicj').text.split(' ')[4][1:4]) #int
    except:
        data['product recommendation'].append(np.nan)
        data['recommendation_count'].append(np.nan) #int
        data['recommendation_percent'].append(np.nan)

    driver.close()
    print("Product", str(c) , data['Product name'][-1] , "--> done")
    print('--------------------')
    time.sleep(50)
```
